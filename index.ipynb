```python
#primero llamamos a pandas y creamos los 4 data frames que tenemos que limpiar
import pandas as pd
Pagina1="https://docs.google.com/spreadsheets/d/1vDV3vr9mX0nmoCJAwIE31Iawoz_0QW9tgTI2yoSCisw/edit?gid=615495222#gid=615495222"
df1=pd.read_html(Pagina1)[0]
df1=df1.iloc[4:]
keep_col=["A","B","C","E","G","I","K","M"]
df1=df1[keep_col]
df1=df1.drop(index=5)
df1=df1.drop(index=6)
df1=df1.drop(index=7)
df1.columns = df1.iloc[0]
df1 = df1[1:]
df1=df1.iloc[:66]
```

```python
#ahora limpiamos el data frame numero 2

Pagina2="https://docs.google.com/spreadsheets/d/1nj3bPlfaSXIzMwncPUaC3-_eZzHSCEs3Bf6eEHMHxiQ/edit?gid=2132261676#gid=2132261676"
df2=pd.read_html(Pagina2)[0]
keep_colum2=["A","B","C","D","E","F","H","J"]
df2=df2[keep_colum2]
df2.columns = df2.iloc[0]
df2 = df2[1:]
df2=df2.drop(index=1)
df2=df2.drop(index=2)
df2=df2.drop(index=3)
df2=df2.iloc[:49]
```

```python
#ahora se corrige el data frame 3
Pagina3="https://docs.google.com/spreadsheets/d/1iqO3LNdn5wjLadxnvh3lDaXj_B_00OC_3uMIysr-Xig/edit?gid=1925950326#gid=1925950326"
df3=pd.read_html(Pagina3)[0]
keep_colum3=["A","B","C","E","G","I","K","M"]
df3=df3[keep_colum3]
df3.columns = df3.iloc[0]
df3 = df3[1:]
df3
df3=df3.drop(index=1)
df3=df3.drop(index=2)
df3=df3.drop(index=3)
df3=df3.iloc[:44]
```

```python
#por ultimo se limpia el data frame 4
Pagina4="https://docs.google.com/spreadsheets/d/1ITiE8ltvfBqPtszpEJcSHFTjL8qLbrj9qkxdBhaW4r4/edit?gid=710188050#gid=710188050"
df4=pd.read_html(Pagina4)[0]
keep_colum4=["A","B","D","E","F","G","H","I"]
df4=df4[keep_colum4]
df4.columns = df4.iloc[0]
df4 = df4[1:]
df4
df4=df4.drop(index=1)
df4=df4.iloc[:32]
```

```python
#una vez todo limpiado, se procede a formar el dataframe1
dataframe1=pd.concat([df1,df2,df3,df4])

dataframe1["HDI rank"]=pd.to_numeric(dataframe1["HDI rank"])
dataframe1["Human Development Index (HDI)"]=pd.to_numeric(dataframe1["Human Development Index (HDI)"])
dataframe1["Life expectancy at birth"]=pd.to_numeric(dataframe1["Life expectancy at birth"])
dataframe1["Expected years of schooling"]=pd.to_numeric(dataframe1["Expected years of schooling"])
dataframe1["Mean years of schooling"]=pd.to_numeric(dataframe1["Mean years of schooling"])
dataframe1["Gross national income (GNI) per capita"]=pd.to_numeric(dataframe1["Gross national income (GNI) per capita"])
dataframe1["GNI per capita rank minus HDI rank"]=pd.to_numeric(dataframe1["GNI per capita rank minus HDI rank"])
dataframe1=dataframe1.rename(columns={"HDI rank": "HDI_rank","Human Development Index (HDI)":"Human_dev_in","Life expectancy at birth":"Life_ex_birth","Expected years of schooling":"Ex_Year_schooling","Mean years of schooling":"Mean_years_schooling","Gross national income (GNI) per capita":"GNI","GNI per capita rank minus HDI rank":"GNI_HDI"})
dataframe1
```

```python
frios_covid=pd.read_csv("fallecidosCOVID.csv",sep=";")
frios_covid
```

```python
frios_covid['FECHA_FALLECIMIENTO'] = frios_covid['FECHA_FALLECIMIENTO'].astype(str)

frios_covid['año_fallecimiento'] = frios_covid['FECHA_FALLECIMIENTO'].str[:4].astype(int)
frios_covid['mes_fallecimiento'] = frios_covid['FECHA_FALLECIMIENTO'].str[4:6].astype(int)
frios_covid['dia_fallecimiento'] = frios_covid['FECHA_FALLECIMIENTO'].str[6:].astype(int)
frios_covid
```

```python
frios_per_month_year = frios_covid.groupby(['año_fallecimiento', 'mes_fallecimiento']).size()
frios_per_month_year = frios_per_month_year.reset_index(name='count')
frios_per_month_year
```

```python
frios_per_month_cause = frios_covid.groupby(['año_fallecimiento', 'mes_fallecimiento', 'CLASIFICACION_DEF']).size()
frios_per_month_cause = frios_per_month_cause.reset_index(name='count')
frios_per_month_cause
```
